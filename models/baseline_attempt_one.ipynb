```python
#Imports
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.inspection import permutation_importance
```

```python
#Data Preprocessing
df = pd.concat([
    pd.read_csv("../data/processed_22.csv"),
    pd.read_csv("../data/processed_23.csv"),
    pd.read_csv("../data/processed_24.csv"),
    pd.read_csv("../data/processed_25.csv")
], ignore_index=True)
df["date"] = pd.to_datetime(df["date"])

y = df["Lpoly_expected_ml"]
X = df.drop(columns=["date", "Lpoly_expected", "Lpoly_expected_ml"])
```

```python
# Linear regression
lin = Pipeline([
    ("scaler", StandardScaler()),
    ("model", LinearRegression())
])
lin.fit(X, y)
lin_pred = lin.predict(X)
```

```python
# Ridge regression
ridge = Pipeline([
    ("scaler", StandardScaler()),
    ("model", Ridge(alpha=1.0))
])
ridge.fit(X, y)
ridge_pred = ridge.predict(X)
```

```python
# Random Forest
rf = RandomForestRegressor(
    n_estimators=200,
    random_state=42,
    n_jobs=-1
)
rf.fit(X, y)
rf_pred = rf.predict(X)
```

```python
results = {
    "Linear_R2": r2_score(y, lin_pred),
    "Linear_MSE": mean_squared_error(y, lin_pred),
    "Ridge_R2": r2_score(y, ridge_pred),
    "Ridge_MSE": mean_squared_error(y, ridge_pred),
    "RF_R2": r2_score(y, rf_pred),
    "RF_MSE": mean_squared_error(y, rf_pred),
}

print("Model Performance Results:")
print("-" * 50)
for key, value in results.items():
    print(f"{key}: {value:.4f}")
```
