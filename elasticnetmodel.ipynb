```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
data_2023 = pd.read_csv('C:/Users/sivar/capstoneB/data/processed_23.csv')
data_2024 = pd.read_csv('C:/Users/sivar/capstoneB/data/processed_24.csv')
data_2025 = pd.read_csv('C:/Users/sivar/capstoneB/data/processed_25.csv')
```

```python
data_2023
```

```python
y_data = data_2023['Lpoly_expected_ml']
x_data = data_2023.drop(columns=['Lpoly_expected_ml'],axis=1)


```

```python
model = ElasticNet(alpha = 0.1, l1_ratio = 0.5)
```

```python
predictions = model.fit(x_data, y_data)
y_pred = model.predict(x_data)
mse = mean_squared_error(y_data, y_pred)
print(f'Mean Squared Error: {mse}')
```

```python
elasticnet_coeff = model.coef_
feature_coeff_df = pd.DataFrame({'Feature': x_data.columns, 'Coefficient': elasticnet_coeff})
feature_coeff_df.drop(feature_coeff_df[feature_coeff_df['Coefficient'] == 0].index, inplace=True)
feature_coeff_df
```
